#Text cleaning

```python
import re

# Function to clean text
def clean_text(text):
    if isinstance(text, str):
        text = text.lower()  # Convert to lowercase
        text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
        text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Apply cleaning to relevant text columns
df['Title'] = df['Title'].apply(clean_text)
df['Author'] = df['Author'].apply(clean_text)

# If the 'Published' column contains year data as text, we clean it too (optional)
df['Published'] = df['Published'].apply(str)  # Convert to string (if not already)
df['Published'] = df['Published'].apply(lambda x: x.strip())  # Remove extra spaces

# Display cleaned data
df.head()
```

#Writing the Recommendation Function

```python
def recommend_books(book_title, similarity_matrix, df, top_n=10):
    """
    Recommend books similar to a given book based on the similarity matrix.

    Returns:
    - List of recommended books with similarity scores.
    """
    if book_title not in df['Title'].values:
        return "Book not found in the dataset!"

    # Find the index of the given book
    book_idx = df[df['Title'] == book_title].index[0]

    # Get similarity scores for all books with the given book
    similarity_scores = list(enumerate(similarity_matrix[book_idx]))

    # Sort books based on similarity scores (excluding the book itself)
    sorted_books = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]

    # Get recommended book titles
    recommended_books = [(df.iloc[i[0]]['Title'], round(i[1], 3)) for i in sorted_books]

    return recommended_books
```

#Collaborative Filtering

###Preparing the Data for Collaborative Filtering

```python
# Creating a "Score" column
df['Score'] = (df['Ratings'] / 5) * np.log1p(df['Shelvings'])
```

```python
df_cf = df[['Title', 'Shelvings', 'Score']].copy()
```

```python
print(df_cf.head())
```

#Constructing User-Book Matrix




```python
# Creating a pivot table (Interaction matrix)
interaction_matrix = df_cf.pivot_table(index='Title', values='Shelvings', aggfunc='sum')

# Normalizing (scaling) the values
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
interaction_matrix_scaled = scaler.fit_transform(interaction_matrix)

# Converting back to DataFrame
interaction_matrix = pd.DataFrame(interaction_matrix_scaled, index=interaction_matrix.index, columns=['Shelvings_Scaled'])

# Checking the processed matrix
print(interaction_matrix.head())
```

#Building the KNN Model for Collaborative Filtering


```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn_model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10)
```

```python
knn_model.fit(interaction_matrix.to_numpy())  # Converts DataFrame to NumPy array
```

```python
# Function to recommend books based on the given book title
def recommend_books_cf(book_title, n_recommendations=5):
    if book_title not in interaction_matrix.index:
        print("Book not found in the dataset.")
        return []

    # Get the index of the given book
    book_index = np.where(interaction_matrix.index == book_title)[0][0]

    # Find the k-nearest neighbors
    distances, indices = knn_model.kneighbors(interaction_matrix.iloc[book_index].values.reshape(1, -1), n_neighbors=n_recommendations + 1)

    # Get recommended books (excluding the first one, which is the input book itself)
    recommended_books = [(interaction_matrix.index[i], 1 - distances[0][j]) for j, i in enumerate(indices[0]) if j != 0]

    return recommended_books
```

```python
book_name = "the english assassin gabriel allon 2"  # Example book
recommendations = recommend_books_cf(book_name)
```

```python
print(f"Recommended books similar to '{book_name}':")
for idx, (title, similarity) in enumerate(recommendations, 1):
    print(f"{idx}. {title} (Similarity Score: {similarity:.4f})")
```

#Hybrid Model - Combining Content-Based & Collaborative Filtering

```python
def hybrid_recommend(book_name, content_weight=0.7, collab_weight=0.3, n_recommendations=10):
    # Get content-based recommendations
    content_recs = content_based_recommend(book_name, n_recommendations)
    content_recs = {title: score for title, score in content_recs}  # Convert to dictionary

    # Get collaborative recommendations
    collab_recs = collaborative_recommend(book_name, n_recommendations)
    collab_recs = {title: score for title, score in collab_recs}  # Convert to dictionary

    # Combine both recommendations
    combined_recs = {}
    for title in set(content_recs.keys()).union(collab_recs.keys()):
        combined_recs[title] = content_recs.get(title, 0) * content_weight + collab_recs.get(title, 0) * collab_weight

    # Sort recommendations based on weighted score
    sorted_recommendations = sorted(combined_recs.items(), key=lambda x: x[1], reverse=True)

    # Print final hybrid recommendations
    print(f"\n🔹 **Hybrid Recommendations for '{book_name}':**")
    for idx, (title, score) in enumerate(sorted_recommendations[:n_recommendations], 1):
        print(f"{idx}. {title} (Hybrid Score: {score:.4f})")

    return sorted_recommendations[:n_recommendations]
```

```python
def content_based_recommend(book_name, n_recommendations=10):
    if book_name not in df['Title'].values:
        print(f" Book '{book_name}' not found in dataset!")
        return []

    # Get book index
    book_idx = df[df['Title'] == book_name].index[0]

    # Get cosine similarity scores for the book
    sim_scores = list(enumerate(cosine_sim[book_idx]))

    # Sort by similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:n_recommendations+1]

    # Get book titles
    recommended_books = [(df.iloc[i[0]]['Title'], i[1]) for i in sim_scores]

    return recommended_books
```

```python
def collaborative_recommend(book_name, n_recommendations=10):
    if book_name not in df['Title'].values:
        print(f" Book '{book_name}' not found in dataset!")
        return []

    # Get book index
    book_idx = df[df['Title'] == book_name].index[0]

    # Find k nearest neighbors
    distances, indices = knn_model.kneighbors(interaction_matrix_scaled[book_idx].reshape(1, -1), n_neighbors=n_recommendations+1)

    # Get recommended books
    recommended_books = [(df.iloc[idx]['Title'], 1 - dist) for idx, dist in zip(indices[0][1:], distances[0][1:])]

    return recommended_books
```

#Adding Popularity-Based Filtering

```python
def popular_books(n=10, min_shelvings=50000, min_ratings=4.0):
    popular = df[(df["Shelvings"] >= min_shelvings) & (df["Ratings"] >= min_ratings)]
    return popular.sort_values(by="Shelvings", ascending=False).head(n)[["Title", "Shelvings", "Ratings"]]
```
